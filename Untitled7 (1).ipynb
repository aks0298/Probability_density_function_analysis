```python
# === Libraries ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.stats import gaussian_kde
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

# === Reproducibility ===
np.random.seed(42)
torch.manual_seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# === Data Preparation ===
def load_and_prepare(path, roll_no):
    data = pd.read_csv(path, encoding="latin1")
    values = data[['no2']].dropna().values.astype(np.float32)

    scaler = StandardScaler()
    scaled = scaler.fit_transform(values).flatten()

    a_val = 0.5 * (roll_no % 7)
    b_val = 0.3 * ((roll_no % 5) + 1)

    transformed = scaled + a_val * np.sin(b_val * scaled)
    return transformed[:8000], a_val, b_val


z_data, a_r, b_r = load_and_prepare(
    "data.csv",
    roll_no=102317084
)

print("\nParameters Used:")
print(f"a_r = {a_r}, b_r = {b_r}")

tensor_data = torch.tensor(z_data, dtype=torch.float32).view(-1, 1)
train_loader = DataLoader(TensorDataset(tensor_data), batch_size=32, shuffle=True)

# === Model Definitions ===
class GenNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(1, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 1)
        )

    def forward(self, x):
        return self.net(x)


class DiscNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(1, 32),
            nn.LeakyReLU(0.2),
            nn.Linear(32, 16),
            nn.LeakyReLU(0.2),
            nn.Linear(16, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.net(x)


G = GenNet().to(device)
D = DiscNet().to(device)

loss_fn = nn.BCELoss()
opt_G = torch.optim.Adam(G.parameters(), lr=3e-4)
opt_D = torch.optim.Adam(D.parameters(), lr=3e-4)

# === GAN Training ===
def train_gan(generator, discriminator, loader, epochs=500):
    for ep in range(epochs):
        for idx, (real_batch,) in enumerate(loader):
            real_batch = real_batch.to(device)
            bs = real_batch.size(0)

            # --- Train Discriminator ---
            noise = torch.randn(bs, 1).to(device)
            fake_batch = generator(noise).detach()

            real_lbl = torch.ones(bs, 1).to(device)
            fake_lbl = torch.zeros(bs, 1).to(device)

            d_loss = (
                loss_fn(discriminator(real_batch), real_lbl) +
                loss_fn(discriminator(fake_batch), fake_lbl)
            )

            opt_D.zero_grad()
            d_loss.backward()
            opt_D.step()

            # --- Train Generator ---
            if idx % 2 == 0:
                noise = torch.randn(bs, 1).to(device)
                gen_samples = generator(noise)
                g_loss = loss_fn(discriminator(gen_samples), real_lbl)

                opt_G.zero_grad()
                g_loss.backward()
                opt_G.step()

        if ep % 50 == 0:
            print(f"Epoch {ep}/{epochs} | D Loss: {d_loss.item():.4f} | G Loss: {g_loss.item():.4f}")

train_gan(G, D, train_loader)
print("\nTraining Finished")

# === Sampling from Generator ===
with torch.no_grad():
    samples = G(torch.randn(3000, 1).to(device)).cpu().numpy().flatten()

# === Density Estimation ===
density = gaussian_kde(samples)
x_axis = np.linspace(samples.min(), samples.max(), 500)
kde_vals = density(x_axis)

hist_vals, edges = np.histogram(samples, bins=60, density=True)
centers = (edges[:-1] + edges[1:]) / 2

# === Visualization ===
plt.figure(figsize=(10, 6))
plt.plot(x_axis, kde_vals, linewidth=2, label="KDE from GAN samples")
plt.plot(centers, hist_vals, '--', label="Histogram estimate")
plt.title("PDF Approximation of Transformed Variable z")
plt.xlabel("z")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()

# === Notes ===
print("\nNotes:")
print(" The generator approximates the multimodal structure induced by the sinusoidal transformation.")
print(" Loss curves indicate stable adversarial learning without collapse.")
print(" KDE and histogram show consistent estimation of the learned distribution.")
```
